```python
from bs4 import BeautifulSoup
from urllib.request import Request, urlopen

site = "http://pwr.nu.edu.pk/computer-science/"
hdr = {'User-Agent': 'Mozilla/5.0'}
req = Request(site, headers=hdr)
page = urlopen(req)
data = BeautifulSoup(page, 'html.parser')
print(data)
```

```python
title = data.title
print(title.text)
```

```python
data_txt = data.get_text()
print(data_txt)
```

```python
E_C = data.find_all('ul', attrs={'class', 'list theme-colored'})
E_C
```

```python


for li in E_C:
    li_txt = li.find_all('li')
print(type(li_txt))

E_C_txt = []
for div in li_txt:
    E_C_txt.append(div.text)
    # print(div.text)

print(E_C_txt)
```

```python
CS_dep = data.find('h2', attrs={'class', 'title text-white'})
CS_dep = CS_dep.text
```

```python
scrapped_data = []
dic = {}

dic['Dep_Name'] = CS_dep
dic['Dep_Eligible_Criteria'] = E_C_txt

scrapped_data.append(dic)

```

```python
import pandas as pd
dataFrame = pd.DataFrame.from_dict(scrapped_data)

print(dataFrame)
```
